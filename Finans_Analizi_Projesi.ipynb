```python
!pip install yfinance
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hisseler = ['THYAO.IS', 'PGSUS.IS']
veri = yf.download(hisseler, start="2025-01-01", end="2026-02-20")['Close']

print(veri.head())
```

```python
gunluk_getiri = veri.pct_change()

print("Ortalama Günlük Getiri:")
print(gunluk_getiri.mean())

print("\nRisk (Standart Sapma):")
print(gunluk_getiri.std())
```

```python
plt.figure(figsize=(10, 5))
plt.plot(veri['THYAO.IS'], label='Türk Hava Yolları')
plt.plot(veri['PGSUS.IS'], label='Pegasus')
plt.title('THY vs Pegasus Fiyat Karşılaştırması')
plt.legend()
plt.show()

gunluk_getiri.hist(bins=50, figsize=(10, 5))
plt.title('Günlük Getiri Dağılımı')
plt.show()
```

```python
korelasyon = gunluk_getiri.corr()

print("THY ve Pegasus Arasındaki Korelasyon:")
print(korelasyon)

import seaborn as sns
plt.figure(figsize=(6, 4))
sns.heatmap(korelasyon, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Hisse Senedi Korelasyon Matrisi')
plt.show()
```
